```python
# Display data types for each attribute
for column in ds.columns:
    data_type = ds[column].dtype
    print(f"Data type for {column}: {data_type}")
```

```python
ds = ds.mask(ds.eq('UNKNOWN'))
# Drop rows with missing values
ds.dropna(inplace=True)
ds.shape
```

```python
ds = ds.mask(ds.eq('UNKNOWN RACE'))
ds.dropna(inplace=True)
ds.shape
```

```python
ds = ds.mask(ds.eq('CLIENT DID NOT ANSWER'))
ds.dropna(inplace=True)
ds.shape
```

```python
ds = ds.mask(ds.eq('UNKNOWN EMPLOYMENT HOURS'))
ds.dropna(inplace=True)
ds.shape
```

```python
from sklearn.preprocessing import LabelEncoder
# Identify categorical columns
categorical_columns = ds.select_dtypes(include=['object']).columns

# Apply label encoding to each categorical column
label_encoder = LabelEncoder()
for column in categorical_columns:
    ds[column] = label_encoder.fit_transform(ds[column])

ds.head()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as pb
# Identify features (X) and target (y)
X = ds.drop('Obesity', axis=1)  # 'Obesity' is the target column
y = ds['Obesity']

# Identify categorical columns and apply label encoding
categorical_columns = X.select_dtypes(include=['object']).columns
for column in categorical_columns:
    X[column] = ps.factorize(X[column])[0]

# Feature Selection - Selecting top 20 features using ANOVA F-statistic
X_top20 = SelectKBest(f_classif, k=20).fit_transform(X, y)

# Get the scores of the features
feature_scores = SelectKBest(f_classif, k=20).fit(X, y).scores_

# Create a DataFrame for visualization
feature_importance_ds = ps.DataFrame({'Feature': X.columns, 'Score': feature_scores})
feature_importance_ds = feature_importance_ds.sort_values(by='Score', ascending=False).head(20)

# Plot the top 20 features
pb.figure(figsize=(12, 8))
pb.barh(feature_importance_ds['Feature'], feature_importance_ds['Score'])
pb.xlabel('F-statistic Score')
pb.title('Top 20 Features based on F-statistic')
pb.show()
```

```python
# Feature Selection - Selecting top 10 features using ANOVA F-statistic
X_top10 = SelectKBest(f_classif, k=10).fit_transform(X, y)
```

```python
# Display selected features
selected_features = X.columns[SelectKBest(f_classif, k=10).fit(X, y).get_support()]
print("Selected Features:", selected_features)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_top10, y, test_size=0.2, random_state=42)

# Decision Tree
dt_classifier = DecisionTreeClassifier()
dt_params = {'max_depth': [None, 10, 20, 30], 'min_samples_split': [2, 5, 10]}
dt_grid = GridSearchCV(dt_classifier, dt_params, cv=5)
dt_grid.fit(X_train, y_train)
```

```python
# Random Forest
rf_classifier = RandomForestClassifier()
rf_params = {'n_estimators': [50, 100], 'max_depth': [None, 10, 20, 30], 'min_samples_split': [2, 5, 10]}
rf_grid = GridSearchCV(rf_classifier, rf_params, cv=5)
rf_grid.fit(X_train, y_train)
```

```python
# SVM
svm_classifier = SVC()
svm_params = {'C': [0.1, 1], 'kernel': ['linear', 'rbf']}
svm_grid = GridSearchCV(svm_classifier, svm_params, cv=5)
svm_grid.fit(X_train, y_train)
```

```python
# Display best parameters
print("Best Parameters - Decision Tree:", dt_grid.best_params_)
print("Best Parameters - Random Forest:", rf_grid.best_params_)
print("Best Parameters - SVM:", svm_grid.best_params_)
```

```python
# Evaluate models
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    return accuracy, precision, recall, f1

# Evaluate Decision Tree
dt_accuracy, dt_precision, dt_recall, dt_f1 = evaluate_model(dt_grid, X_test, y_test)

# Evaluate Random Forest
rf_accuracy, rf_precision, rf_recall, rf_f1 = evaluate_model(rf_grid, X_test, y_test)

# Evaluate SVM
svm_accuracy, svm_precision, svm_recall, svm_f1 = evaluate_model(svm_grid, X_test, y_test)

# Print results
print("Decision Tree - Accuracy: {}, Precision: {}, Recall: {}, F1 Score: {}".format(dt_accuracy, dt_precision, dt_recall, dt_f1))
print("Random Forest - Accuracy: {}, Precision: {}, Recall: {}, F1 Score: {}".format(rf_accuracy, rf_precision, rf_recall, rf_f1))
print("SVM - Accuracy: {}, Precision: {}, Recall: {}, F1 Score: {}".format(svm_accuracy, svm_precision, svm_recall, svm_f1))
```

```python
# Results
models = ['Decision Tree', 'Random Forest', 'SVM']
accuracy = [dt_accuracy, rf_accuracy, svm_accuracy]
precision = [dt_precision, rf_precision, svm_precision]
recall = [dt_recall, rf_recall, svm_recall]
f1 = [dt_f1, rf_f1, svm_f1]

# Bar plot
fig, ax = pb.subplots(figsize=(10, 6))

# Plotting metrics for each model
bar_width = 0.2
index = range(len(models))
bar1 = ax.bar(index, accuracy, width=bar_width, label='Accuracy')
bar2 = ax.bar([i + bar_width for i in index], precision, width=bar_width, label='Precision')
bar3 = ax.bar([i + 2 * bar_width for i in index], recall, width=bar_width, label='Recall')
bar4 = ax.bar([i + 3 * bar_width for i in index], f1, width=bar_width, label='F1 Score')

# Labeling
ax.set_xlabel('Models', fontsize=14)
ax.set_ylabel('Metrics', fontsize=14)
ax.set_title('Model Performance Comparison', fontsize=16)
ax.set_xticks([i + 1.5 * bar_width for i in index])
ax.set_xticklabels(models)
ax.legend()

# Show the plot
pb.show()
```
